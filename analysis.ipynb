```python
import pandas as pd
import numpy as np
from IPython.display import display
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("result.dat",sep=" ", header=None, names=["StartX", "Bin"])
```

```python
df['Bin']=df['Bin'].str.extract(r'(\d+)')
df = df.astype({"Bin": int, "StartX": float})
```

```python
display(df)
print(df.dtypes)
```

```python
df['Bin'].describe()
```

```python
ax = sns.histplot(df.sort_values(by="Bin"), x="Bin", bins=range(1, 20), discrete=True)

for patch in ax.patches:
    height = patch.get_height()
    x = patch.get_x() + patch.get_width() / 2
    y = height
    ax.text(x, y, f'{int(height)}', ha='center', va='bottom')

xticks = ax.get_xticks()
ax.set_xticks(range(1, 20)) 
ax.set_xticklabels(range(1, 20)) 

plt.xlabel('Bin')
plt.ylabel('Frequency')
plt.title('Ball Distribution')
plt.show()
```

```python
ax = sns.scatterplot(df, x="StartX", y="Bin")
ax.set_yticks(range(1, 20))
ax.set_yticklabels(range(1, 20))
plt.show()
```
